```python
def change_date(x):
    return x[0:2] + '14' + x[4:]
```

```python
frames = [p1,p2,p3,p4]
tran_data = pd.concat(frames)
tran_data['RecordedAtTime'] = tran_data['RecordedAtTime'].apply(lambda x: change_date(x))
tran_data.head()
```

```python
la_mean = taxi_data["pickup_latitude"].mean()
lo_mean = taxi_data["pickup_longitude"].mean()
la_std = taxi_data["pickup_latitude"].std()
lo_std = taxi_data["pickup_longitude"].std()
```

#### We will filter out outlier values by CLT, mean +/- 3* std

```python
taxi_data = taxi_data[taxi_data['pickup_latitude'] < (la_mean+3*la_std)]
taxi_data = taxi_data[taxi_data['pickup_latitude'] > (la_mean-3*la_std)]
taxi_data = taxi_data[taxi_data['pickup_longitude'] < (lo_mean+3*lo_std)]
taxi_data = taxi_data[taxi_data['pickup_longitude'] > (lo_mean-3*lo_std)]
```

### Conclusion By Now:
##### The range of uber pickup location is narrow, and most are located between [(40.5 to 41.2),(-74.5 to -73.2)]. The range of taxi pickup location is much wider, and most are located between [(36 to 45.5),(-82.5 to -65.4)]. The range for public transport is more like city map, we could see which part are more busy than other part from the visulization of public transport scatter map.
##### From the usage/number of pickup during each week, we could find that more taxi and uber pickups during Tue, Wed and Thur. More pickups are done during AM timezone for both taxi and uber. People are more likely to choose uber in the cneter area, for those they do not choose public transport. People will choose taxi out of center area, since there is no enough ubers and public transport.

```python
processed_taxi['WeekDayName'] = processed_taxi['WeekDayName'].astype('category')
processed_taxi['WeekDayName'] = processed_taxi['WeekDayName'].cat.codes
processed_taxi['Period'] = processed_taxi['Period'].astype('category')
processed_taxi['Period'] = processed_taxi['Period'].cat.codes
```

```python
X = processed_taxi.drop(['total_amount'],axis=1)
y = processed_taxi['total_amount']
reg = LinearRegression().fit(X, y)
reg.coef_
```

#### We could see, after we use the multi regression model to fit the taxi dataset, Month, Period and Weekday are the top 3 most important factors.

#### From the previous part, the relation between location, Weather and Taxi trip duration, we know weather is the most important factor that affect the taxi drivers. The change of month means the different pattern in different months. Period and Weekday are the second and third important factors, which is as same as we conclude in the previous part.

```python
from sklearn.cluster import KMeans
num_clusters = 10
X = uber_friday[['Lat','Lon']].to_numpy()
kms = KMeans(n_clusters = num_clusters)
y = kms.fit_predict(X)
#plt.scatter(uber_friday['Lat'], uber_friday['Lon'], c=y)
#plt.show()
#print(kms.cluster_centers_)
centers = kms.cluster_centers_
cluster_count = [0]*num_clusters
for item in y:
    cluster_count[item] +=1
cluster_max=0
max_index = 0
for i in range(len(cluster_count)):
    if cluster_count[i] > cluster_max:
        cluster_max = cluster_count[i]
        max_index = i
    cluster_count[i] = cluster_count[i]/100
print(centers[max_index])
print(cluster_max)
plt.scatter(centers[:,0], centers[:,1], s= cluster_count, marker='o')
plt.show()
```
